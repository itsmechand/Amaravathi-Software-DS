**1)Plot the count of males and females in the data set.**

```python
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sbn
dsd = pd.read_csv('Diwali Sales Data.csv',encoding='ISO- 8859-1')
pt.figure(figsize=(8, 6))
sbn.countplot(data=dsd, x='Gender')
pt.title('Count of Males and Females')
pt.xlabel('Gender')
pt.ylabel('Count')
pt.show()
```

**2. Give the sum of amounts spent by each gender and plot the corresponding graph.**


```python
spending_by_gender = dsd.groupby('Gender')['Amount'].sum()
pt.figure(figsize=(8, 6))
spending_by_gender.plot(kind='bar',color='maroon')
pt.title('Total Amount Spent by Each Gender')
pt.xlabel('Gender')
pt.ylabel('Total Amount Spent')
pt.show()
```

**3. Count each age group and provide individual counts grouped by gender.**


```python
agc = dsd.groupby(['Age Group','Gender']).size().unstack()
print(agc)
```

**4. Plot the total amount spent by each age group.**


```python
ags = dsd.groupby('Age Group')['Amount'].sum()
pt.figure(figsize=(8, 6))
ags.plot(kind='bar',color='orange')
pt.title('Total Amount spent by each Age Group')
pt.xlabel('Age Group')
pt.ylabel('Total Amount Spent')
pt.show()
```

**5. Plot a graph depicting the total number of orders from the top 10 states.**


```python
top_states = dsd['State'].value_counts().head(10)
pt.figure(figsize=(12, 6))
top_states.plot(kind='bar',color='navy')
pt.title('Total No.of Orders from Top 10 States')
pt.xlabel('State')
pt.ylabel('Number of Orders')
pt.show()
```

**6. Determine the total amount spent in the top 10 states.**


```python
top_ten_states = dsd[dsd['State'].isin(top_states.index)].groupby('State')['Amount']. sum()
pt.figure(figsize=(12, 5))
top_ten_states.plot(kind='bar',color='green')
pt.title('Total Amount Spent in Top 10 States')
pt.xlabel('State')
pt.ylabel('Total Amount Spent')
pt.show()
```

**7. Plot a comparison graph between the number of married and unmarried individuals.**


```python
pt.figure(figsize=(8, 6))
sbn.countplot(data=dsd, x='Marital_Status')
pt.title('Comparison of Married and Unmarried Individuals')
pt.xlabel('Marital Status')
pt.ylabel('Count')
pt.show()
```

**8. Plot the amount spent by males and females based on marital status.**


```python
marital_gender_amounts = dsd.groupby(['Marital_Status','Gender'])['Amount'].sum().unstack()
pt.figure(figsize=(10, 6))
marital_gender_amounts.plot(kind='bar')
pt.title('Amount Spent by Males and Females Based on Marital Status')
pt.xlabel('Marital Status')
pt.ylabel('Total Amount Spent')
pt.show()
```

**9. Plot the count of each occupation present in the dataset.**


```python
occupation_counts = dsd['Occupation'].value_counts()
pt.figure(figsize=(12, 4))
occupation_counts.plot(kind='bar', color='tomato')
pt.title('Count of Each Occupation')
pt.xlabel('Occupation')
pt.ylabel('Count')
pt.show()
```

**10. Plot the amount spent by each occupation in descending order.**


```python
occupation_amounts = dsd.groupby('Occupation')['Amount'].sum().sort_values(ascending=False)
pt.figure(figsize=(12, 4))
occupation_amounts.plot(kind='bar', color='aqua')
pt.title('Amount Spent by Each Occupation in Descending Order')
pt.xlabel('Occupation')
pt.ylabel('Total Amount Spent')
pt.show()
```

**11. Provide statistical analysis of each product category based on the percentage of orders completed.**


```python
category_based_orders = dsd.groupby('Product_Category')['Orders'].describe()
print(category_based_orders)
```

**12. Determine the budget spent on each product category in descending order.**


```python
category_based_spending = dsd.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
pt.figure(figsize=(13, 5))
category_based_spending.plot(kind='bar', color='purple')
pt.title('Total Amount Spent on Each Product Category')
pt.xlabel('Product Category')
pt.ylabel('Total Amount Spent')
pt.show()
```
